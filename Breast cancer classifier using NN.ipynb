### Importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# splitting data into train & test
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

### Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
X_train_std
```

### Building the Neural Network

- Tensorflow is a Deep Learning library developed by google
- Other libraries for building NNs are Pytorch, Microsoft cntk (Cognition toolkit)
- keras is not a standalone library, just a wrapper of tensorflow
- keras made building NNs using tensorflow easier by creating several APIs and Functions.
- so we set our backend working to tensorflow using keras.

```python
# importing tensorflow & keras
import tensorflow as tf
tf.random.set_seed(3)              # random seed is similar to random state
from tensorflow import keras
```

```python
# setting up the layers of NN
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),  # input layer
    keras.layers.Dense(20, activation = 'relu'),    # hidden layer
    keras.layers.Dense(2, activation = 'sigmoid')    # output layer
])
```

- Dense() implies that all the neurons in the particular layer is connected with all the neurons in the previous layer.
- When working with NNs we need to feed them input as **single dimensional array**, that's what Flatten() function helps us to do.
- relu stands for rectified linear unit.
- sigmoid is similar to logistic regresion function.
- Basic idea:
- There are 30 features. so for each feature we will give one neuron, so by that there will be 30 neurns in input layer.
- **Note:** This is specific for **numerical dataset** alone, image dataset follows another procedure.\
- Output neurons number accounts on the no.of classes we have. (i.e, label).
- when using target variable values as 0,1,2 and so on, we use **sparse_categorical_crossentropy**.
- when we have one hot encoded labels then we will use **categorical_crossentropy**.

```python
# compiling the NNs
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
# training the NN
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 10)
```

### Visualizing accuracy & loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(['Training data', 'Validation data'], loc = 'lower right')
# note during plotting epoch's index start from 0 not from 1. (i.e, 0 to 9 instead of 1 to 10)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(['Training data', 'Validation data'], loc = 'upper right')
```

### Accuracy of model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

### Buidling predictive system

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
# find labels [0 or 1] for all data points
# model.predict() gives the prediction probability of each class for that data point
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0]) # gives two values(i.e, confidence probabilties of being 0 and 1) for one data point
```

```python
# argmax function example
my_list = [10,20,30]
my_list2 = [0.56,0.23]
index_of_max_value = np.argmax(my_list)
index_of_max_value2 = np.argmax(my_list2)
print(my_list)
print(index_of_max_value)
print(my_list2)
print(index_of_max_value2)
```

**argmax()** helps to find the index value of the **max value** within the array.

```python
# converting prediction probilities to class labels
Y_pred_labels = [np.argmax(i) for i in Y_pred]  # for loop iterates through all 114 data points
print(Y_pred_labels)
```

```python
# predictive system
input_data = (20.57,0,0,17.77,132.9,1326,0.00474,0.07864,0.0069,0.07017,0.1812,0.05667,0.5435,0.7339,3.398,74.08,0.005225,0.01300,0.0106,0.0134,0.01389,0.003532,24.99,23.41,158.0,1956,0.1238,0.1866,0.275,0.08902)

# change the input data into numpy array
input_data_array = np.asarray(input_data)
# reshape the array
input_reshaped = input_data_array.reshape(1,-1)
# standardizing the input data
input_std = scaler.transform(input_reshaped)

prediction = model.predict(input_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
    print("The tumor is malignant")
else:
    print("The tumor is benign")
```
